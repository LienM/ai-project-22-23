# ZIP Code Hashing

I hoped that h&m had either forgotten to pad the zip code hashes in the customers table, or used a very simple padding, which would have allowed brute forcing the table by hashing a bunch of zip codes.

The methodology was to assume that the ZIP code with the most customers was actually a NULL value, and to try hashing various variations on NULL using common cryptographic hash algorithms that return hashes of the right length. This did not work out so I never mentioned it in any presentations.

```python
def padLeft(to_pad, character, amount):
    return amount * character + to_pad
```

```python
def padRight(to_pad, character, amount):
    return to_pad + amount * character
```

```python
import string
import copy
characters = list(string.ascii_uppercase + string.ascii_lowercase)
start_words = set([""])
previous_set = set([""])
for i in range(1, 5):
    temp_set = set()
    for prefix in previous_set:
        for chara in characters:
            temp_set.add(prefix + chara)
            start_words.add(prefix + chara)
    previous_set = temp_set
start_words
```

```python
start_words = ["NaN", "Nan", "NAN", "nan", "None", "none", "NONE", '-1', '0', 'na', 'NA', 'Na', '', "NULL", "Null", "null"]
```

```python
result_words = [st for st in start_words]
```

```python
for i in range(20):
    for word in start_words:
        result_words.append(padLeft(word, " ", i))
        result_words.append(padLeft(word, "0", i))
        result_words.append(padRight(word, " ", i))
        result_words.append(padRight(word, "0", i))
```

```python
import hashlib
```

```python
hashes = dict()
for word in result_words:
    hashes[hashlib.sha256(str.encode(word)).hexdigest()] = word
```
